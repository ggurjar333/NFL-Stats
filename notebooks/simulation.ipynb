### Import libraries

```python
from src.sportsradar.simulation.gamefeeds import GameFeeds 
from src.sportsradar.simulation.gamefeeds import get_game_boxscore, get_game_info, get_game_roster, get_pbp_info
```

### Extract recording_id and session_id

```python
game_feeds = GameFeeds()
rec_id = game_feeds.get_available_recordings()
session_id = game_feeds.get_session(recording_id=rec_id)
game_feeds_data = {'rec_id': rec_id, 'session_id': session_id}
print(game_feeds_data)
```

### Extract boxscore

```python
game_boxscore = get_game_boxscore(recording_id=game_feeds_data['rec_id'], session_id=game_feeds_data['session_id'])
print(game_boxscore)
```

### Extract game info

```python
game_info = get_game_info(recording_id=game_feeds_data['rec_id'], session_id=game_feeds_data['session_id'])
print(game_info)
```

### Extract game roster

```python
game_rosters = get_game_roster(recording_id=game_feeds_data['rec_id'], session_id=game_feeds_data['session_id'])
print(game_rosters)
```

### Extract game play-by-play

```python
game_pbp_info = get_pbp_info(recording_id=game_feeds_data['rec_id'], session_id=game_feeds_data['session_id'])
print(game_pbp_info)
```
